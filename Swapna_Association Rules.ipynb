# Association Rules for Movies Data

```python
import pandas as pd
!pip install mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv('my_movies.csv')
df
```

```python
new_df=df.iloc[:,5:]
new_df
```

```python
itemslist=apriori(new_df,min_support=0.2,use_colnames=True)
itemslist
```

```python
movie_rules=association_rules(itemslist,metric='lift',min_threshold=0.7)
movie_rules
```

```python
movie_rules.sort_values(by='lift',ascending=False)
```

```python
plt.scatter(movie_rules['support'],movie_rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Scatter Plot between Support and Confidence')
```

```python
import seaborn as sns
```

```python
movie_df=pd.DataFrame({'Lift':movie_rules['lift'],'Conviction':movie_rules['conviction']})
movie_df
```

```python
sns.heatmap(movie_df,cmap='Greys',annot=True)
```
