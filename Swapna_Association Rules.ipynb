# Association Rules for Movies Data

```python
import pandas as pd
!pip install mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv('my_movies.csv')
df
```

```python
new_df=df.iloc[:,5:]
new_df
```

```python
itemslist=apriori(new_df,min_support=0.2,use_colnames=True)
itemslist
```

```python
movie_rules=association_rules(itemslist,metric='lift',min_threshold=0.7)
movie_rules
```

```python
movie_rules.sort_values(by='lift',ascending=False)
```

```python
plt.scatter(movie_rules['support'],movie_rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.grid()
plt.title('Scatter Plot between Support and Confidence')
```

```python
import seaborn as sns
```

```python
movie_df=pd.DataFrame({'Lift':movie_rules['lift'],'Conviction':movie_rules['conviction'],'support':movie_rules['support']})
movie_df
```

```python
support_table=movie_rules.pivot(index='antecedents',columns='consequents',values='support')
support_table
```

```python
sns.heatmap(support_table,annot=True,cmap=plt.cm.Blues)
```

```python
##1. Above is a heatmap for support values of all antecendents and consequents
##2. Heatmap above show the support values for all the rules we generated.
## Lets try the heatmap with lift values
```

```python
table_lift=movie_rules.pivot(index='antecedents',columns='consequents',values='lift')
table_lift
```

```python
sns.heatmap(table_lift,annot=True,cmap=plt.cm.Accent_r)
##from the below visualization it is evident that most of lift values
##are below 1.5 and only two values are in range of 4.5 to 5 and 1.5 to 2.0.
```

```python
##get items with support of 0.3
itemslist1=apriori(new_df,min_support=0.3,use_colnames=True)
itemslist1
```

```python
rules1=association_rules(itemslist1,metric='support',min_threshold=0.5)
rules1
```

```python
plt.scatter(rules1['support'],rules1['confidence'],alpha=0.8,c='purple')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.grid()
plt.title('Scatter plot between Support and confidence')
```

```python
##lets build a heatmap with confidence values
rules1_table=rules1.pivot(index='antecedents',columns='consequents',values='confidence')
rules1_table
```
