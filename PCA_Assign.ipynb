# K Means Clustering on PCA Components

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_data=pd.read_csv("wine.csv")
k_data
```

```python
##finding K Value
from sklearn.cluster import KMeans
WCSS=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    model = kmeans.fit(k_data)
    WCSS.append(model.inertia_)
    
```

```python
## Plotting Elbow Curve
plt.plot(range(1,10),WCSS,label='Elbow Curve',c='purple')
plt.xlabel('K Value')
plt.ylabel('WCSS Values')
plt.title('Elbow Curve')
plt.legend()
```

# Working with University Data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
univ_data=pd.read_csv('Universities_Clustering.csv')
univ_data
```

```python
univ_data.describe()
```

```python
univ_data.columns
```

```python
univ_data.shape
```

```python
univ_data.dtypes
```

```python
univ_df=univ_data.drop("Univ",axis=1)
univ_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
univ_df_norm=scale.fit_transform(univ_df)
univ_df_norm=pd.DataFrame(univ_df_norm)
univ_df_norm
```

### Hierarchial Clustering on Original Data

```python
link_univ_orig=linkage(univ_df_norm,method='single',metric='euclidean')
link_univ_orig
dend_orig=dendrogram(link_univ_orig)
dend_orig
```

```python
Aggcl_orig=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='single').fit(univ_df_norm)
Aggcl_orig.labels_
```

```python
univ_data['orig_labels']=Aggcl_orig.labels_
```

###KMeans Clustering on Original Data

```python

WCSS=[]
for i in range(1,10):
    k_orig=KMeans(n_clusters=i)
    k_orig=k_orig.fit(univ_df_norm)
    WCSS.append(k_orig.inertia_)
plt.plot(range(1,10),WCSS)
```

```python
k_orig=KMeans(n_clusters=3)
k_orig=k_orig.fit(univ_df_norm)
k_orig.labels_
```

```python
univ_data['Orig_K_Labels']=k_orig.labels_
univ_data
```

### PCA Analysis

```python
##select component as 7 (n+1)
from sklearn.decomposition import PCA
pca=PCA(n_components=6)
pca_components=pca.fit_transform(univ_df_norm)
pca_components
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
np.sort(var)
var
```

```python
var=var*100
var
```

```python
var=np.cumsum(var)
var
```

```python
plt.plot(var,c='Yellow')
```

```python
##Converting PCA components to Dataframe
pca_scores=pd.DataFrame(pca_components,columns=['PCA1','PCA2','PCA3','PCA4','PCA5','PCA6'])
pca_scores
```

```python
##Considering first four PCA components for Clustering
pca_univ=pd.DataFrame(pca_scores.iloc[:,:4])
pca_univ

##for three PCA Components clustering labels are different for original and pca, so performing clustering using 4 pca components'
```

```python
##Perform Hierarchial Clustering on PCA Components
```

```python
link_univ=linkage(pca_univ,method='single',metric='euclidean')
link_univ
```

```python
dend_univ=dendrogram(link_univ)
dend_univ
```

```python
AggCl_Univ=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='single').fit(pca_univ)
AggCl_Univ
```

```python
AggCl_Univ.labels_
```

```python
pca_univ['Labels']=AggCl_Univ.labels_
pca_univ
```

### Comparing PCA Scores with cluster labels

```python
import seaborn as sns
colours=['Brown','Green','Yellow','Blue']
sns.scatterplot(data=pca_univ,x='PCA1',y='PCA2',hue='Labels',palette=colours)
```

```python
univ_data['pca_hie_labels']=AggCl_Univ.labels_
univ_data
```

```python
##K Means Clustering on PCA Data
WCSS=[]
for i in range(1,10):
    kmeans_univ=KMeans(n_clusters=i)
    kmeans_univ=kmeans_univ.fit(pca_univ)
    WCSS.append(kmeans_univ.inertia_)

plt.plot(range(1,10),WCSS)
```

```python
##consider 3 clusters
k_means_univ=KMeans(n_clusters=3)
k_means_univ=k_means_univ.fit(pca_univ)
```

```python
k_means_univ.labels_
```

```python
univ_data['K_labels_pca']=k_means_univ.labels_
```
